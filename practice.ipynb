```python
x = tf.constant(np.array([np.random.randint(-1000,1000,size=10),np.random.randint(-1000,1000,size=10)]))
x
```

```python
x.shape, x.ndim, tf.size(x)
```

```python
tf.reduce_max(x), tf.reduce_min(x), tf.reduce_mean(x), tf.cast(tf.reduce_sum(x)/tf.size(x),dtype=tf.int32)
```

```python
# Variance
y = [j for i in x for j in i]
var = tf.math.reduce_sum([tf.math.pow(i - tf.reduce_mean(x), 2) for i in y])
tf.math.sqrt(var/tf.size(x)), tf.math.reduce_std(tf.cast(x,tf.float64))
```
